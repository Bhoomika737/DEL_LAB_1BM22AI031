```python
import pandas as pd
import numpy as np
df=pd.read_csv('/content/breast-cancer.csv')
df.head()
```

```python
def tn(x):
  return np.tanh(x)

def sigmoid(x):
  return (1/(1+np.exp(-x)))

df=df.sample(frac=1,random_state=42)
df=df.iloc[:20,:6]
df=df.drop(columns='id')
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix

df['diagnosis'] = df['diagnosis'].map({'B': 0, 'M': 1})

features = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean']
X = df[features].values
y = df['diagnosis'].values

s = StandardScaler()
X_scaled = s.fit_transform(X)
weights = np.random.rand(X_scaled.shape[1])
X_scaled=tn(X_scaled)

linear_output = np.dot(X_scaled, weights)
predictions=sigmoid(linear_output)

predictions_binary = np.where(predictions >= 0.5, 1, 0)
df['predicted'] = predictions_binary.flatten()

print(df[['diagnosis', 'predicted']])
print(confusion_matrix(df['predicted'],df['diagnosis']))
```
