```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def step_func(x):
  if x<0:
    return 0
  elif x>=0:
    return 1

def sigmoid(x):
  return 1/1+(np.exp(-x))

def relu(x):
  return max(0,x)

def leaky_relu(x):
  return max(0.01*x,x)

def tn(x):
  return np.tanh(x)

x = np.linspace(-10, 10, 400)
y_step = [step_func(i) for i in x]
y_sigmoid = sigmoid(x)
y_relu = [relu(i) for i in x]
y_leaky_relu = [leaky_relu(i) for i in x]
y_tanh=[tn(i) for i in x]
```

```python

plt.subplot(3,2,2)
plt.title("Step Function")
plt.plot(x, y_step, label='Step', color='blue')
plt.show()

plt.subplot(3,2,3)
plt.title("sigmoid")
plt.plot(x, y_sigmoid, label='Step', color='blue')
plt.show()

plt.subplot(3,2,4)
plt.title("relu")
plt.plot(x, y_relu, label='Step', color='blue')
plt.show()

plt.subplot(3,2,5)
plt.title("leaky relu")
plt.plot(x, y_leaky_relu, label='Step', color='blue')
plt.show()

plt.subplot(3,2,6)
plt.title("tanh relu")
plt.plot(x, y_tanh, label='Step', color='blue')
plt.show()
```
